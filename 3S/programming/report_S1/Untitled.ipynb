```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = np.array([39, 49, 118, 274, 443, 1454, 1612, 1624, 1723, 5300])
num_o = np.array([
    131, 167, 476, 1612, 1623,
    5300, 5824, 6050, 6511, 21842])
num_s = np.array([
    275, 403, 1270, 4982, 4249,
    14110, 15684, 16916, 18485, 60498])
time_1 = np.array([
    0.002, 0.003, 0.026, 0.327, 1.143,
    1*60+1.811, 1*60+25.543, 1*60+29.979, 1*60+15.619, 35*60+56.509])
time_2 = np.array([
    0.001, 0.001, 0.003, 0.009, 0.010,
    0.042, 0.064, 0.065, 0.094, 1.168])
time_3 = np.array([
    0.001, 0.002, 0.005, 0.042, 0.136,
    7.941, 17.853, 20.215, 22.732, 21*60+0.924])
```

```python
plt.scatter(size**3, time_1)
plt.plot(
    size**3, np.poly1d(
        np.polyfit(size**3, time_1, 1))(size**3),
    color='black', alpha=0.5)
plt.xlabel('行列のサイズの3乗')
plt.ylabel('手法①の実行時間[s]')
```

```python
plt.scatter(size**3, time_3)
plt.plot(
    size**3, np.poly1d(
        np.polyfit(size**3, time_3, 1))(size**3),
    color='black', alpha=0.5)
plt.xlabel('行列のサイズの3乗')
plt.ylabel('手法③の実行時間[s]')
```

```python
plt.scatter(size[:9]**3, time_1[:9])
plt.plot(
    size[:9]**3, np.poly1d(
        np.polyfit(size**3, time_1, 1))(size[:9]**3), color='black', alpha=0.5)
plt.xlabel('行列のサイズの3乗')
plt.ylabel('手法①の実行時間[s]')
```

```python
plt.scatter(size[:9]**3, time_3[:9])
plt.plot(
    size[:9]**3, np.poly1d(
        np.polyfit(size**3, time_3, 1))(size[:9]**3), color='black', alpha=0.5)
plt.xlabel('行列のサイズの3乗')
plt.ylabel('手法③の実行時間[s]')
```

```python
plt.scatter(size**2, time_2)
plt.plot(
    size**2, np.poly1d(
        np.polyfit(size**2, time_2, 1))(size**2),
    color='black', alpha=0.5)
plt.xlabel('')
plt.ylabel('手法の実行時間') 
```
