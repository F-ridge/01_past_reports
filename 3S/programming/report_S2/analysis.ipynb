```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter as sf
```

```python
size_dijkstra = np.array([
    10, 50, 100, 300, 500,
    1000, 2000, 3000, 4000, 5000,
    6000, 7000, 8000, 9000
])
size_dp = np.array([
    10, 50, 100, 300, 500,
    1000, 2000, 3000, 4000, 5000,
    6000, 7000, 8000, 9000, 10000,
    20000, 30000, 40000, 50000
])
time_dijkstra = np.array([
    0.02, 0.003, 0.011, 0.053, 0.104,
    0.427, 1.756, 4.161, 8.171, 14.698,
    20.914, 29.923, 42.05, 4*60+49.99
])
time_dp = np.array([
    0.001, 0.002, 0.002, 0.002, 0.004,
    0.012, 0.025, 0.043, 0.068, 0.103,
    0.151, 0.2, 0.256, 0.324, 0.399,
    1.589, 3.694, 6.787, 10.316
])
```

```python
plt.scatter(size_dp*1.1, time_dp, label="動的計画法")
plt.scatter(size_dijkstra*1.1, time_dijkstra, label="ダイクストラ法")
plt.xlabel("文字列の長さの平均")
plt.ylabel("time [s]")
plt.legend()
```

```python
n_dp = size_dp*size_dp*1.2
plt.scatter(n_dp, time_dp / n_dp)
plt.xlabel("$N^2$")
plt.ylabel("time / $N^2$ [s]")
plt.ylim(0, 0.00000002)
ax = plt.gca()
ax.xaxis.set_major_formatter(sf(useMathText=True))
ax.yaxis.set_major_formatter(sf(useMathText=True))
ax.ticklabel_format(style='sci', axis='x', scilimits=(0, 0))
ax.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))
```

```python
plt.scatter(n_dp[9:], time_dp[9:] / n_dp[9:])
plt.xlabel("$N^2$")
plt.ylabel("time / $N^2$ [s]")
plt.ylim(0, 0.0000000037)
ax = plt.gca()
ax.xaxis.set_major_formatter(sf(useMathText=True))
ax.yaxis.set_major_formatter(sf(useMathText=True))
ax.ticklabel_format(style='sci', axis='x', scilimits=(0, 0))
ax.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))
```

```python
n_dijkstra = size_dijkstra*size_dijkstra*1.2
plt.scatter(
    n_dijkstra*np.log(n_dijkstra),
    time_dijkstra / n_dijkstra*np.log(n_dijkstra),
    color='#ff7f0e')
plt.xlabel("$N^2 \log{N^2}$")
plt.ylabel("time / $N^2 \log{N^2}$ [s]")
plt.ylim(0, 0.00006)
ax = plt.gca()
ax.xaxis.set_major_formatter(sf(useMathText=True))
ax.yaxis.set_major_formatter(sf(useMathText=True))
ax.ticklabel_format(style='sci', axis='x', scilimits=(0, 0))
ax.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))
```

```python
plt.scatter(
    n_dijkstra[:-1]*np.log(n_dijkstra[:-1]),
    time_dijkstra[:-1] / n_dijkstra[:-1]*np.log(n_dijkstra[:-1]),
    color='#ff7f0e')
plt.xlabel("$N^2 \log{N^2}$")
plt.ylabel("time / $N^2 \log{N^2}$ [s]")
plt.ylim(0, 0.000011)
ax = plt.gca()
ax.xaxis.set_major_formatter(sf(useMathText=True))
ax.yaxis.set_major_formatter(sf(useMathText=True))
ax.ticklabel_format(style='sci', axis='x', scilimits=(0, 0))
ax.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))
```
